```python
# 비선형 활성화 함수

# 입력을 받아 수학적 변환을 수행하고 출력을 생성하는 함수. ex) 시그모이드, 소프트맥스
# 인공 신경망의 은닉층에서 활성화함수로 시그모이드 함수를 사용하는 것을 지양하는 이유
# 렐루(ReLU) 함수 소개
# 기타 다른 함수 소개

import torch
import matplotlib.pyplot as plt # 맷플롭립 사용
import numpy as np
```

```python
# 활성화 함수의 특징 - 비선형 함수(Nonlinear funtion)

# 활성화 함수로 선형 함수를 사용하게 되면 은닉층을 쌓을 수 없다
# f(x) = W * x 라고 가정
# 은닉층이 2개일 시 출력층을 포함해서 y(x) = f(f(f(x))) 가 된다 -> W * W * W * x -> 선형 함수로는 은닉층을 여러번 추가해도 1회 추가한것과 동일
```

```python
# 시그모이드 함수(Sigmoid funtion) 와 기울기 소실

# 시그모이드 함수 그래프를 그리는 코드
def Sigmoid(x):
    return 1/(1 + np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y = Sigmoid(x)

plt.plot(x, y)
plt.plot([0,0], [1.0, 0.0], ':') # 가운데 점선 추가
plt.title('Sigmoid Function')
plt.show()
```

```python
# 함수의 출력값이 0 또는 1에 가까워질수록 그래프의 기울기가 완만해진다
# 역전파 과정해서 0에 가까운 작은 기울기가 곱해지면 앞에는 기울기가 잘 전달되지 않게 돈다 -> 기울기 소실(Vanishing Gradient)
```

```python
# 하이퍼볼릭탄젠트 함수(Hyperbolic tangent funtion)
```

```python
# 렐루 함수(ReLU)
# 가장 인기있는 활성화 함수. f(x) = max(0, x)

def relu(x):
    return np.maximum(0, x)

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)

plt.plot(x, y)
plt.plot([0,0], [5.0, 0.0], ':') # 가운데 점선 추가
plt.title('ReLU Function')
plt.show()
```

```python
# 렐루는 음수를 입력하면 0을 출력하고, 양수를 입력하면 입력값을 반환
# 특정 양수값에 수렴하지 않으므로 깊은 신경망에서 시그모이드보다 훨씬 더 잘 작동
# 단순 임계값이므로 연산 속도도 빠르다

# 문제점 : 입력값이 음수면 기울기도 0이 된다 -> 이 뉴런은 다시 회생하기 어렵다 : 죽은 렐루(dying ReLU)
```

```python
# 리키 렐루(Leaky ReLU)

# 죽은 렐루를 보완하기 위한 ReLU 의 변형 함수중 하나
# 리키 렐루ㅅ는 입력값이 음수일 때 0 이 아니라 0.001 과 같은 매우 작은 수를 반환하도록 되어있다

# 수식 : f(x) = max(ax, x)
# a : 하이퍼파라메터, leaky('새는') 정도를 결정, 일반적으로 0.01

a = 0.1

def leaky_relu(x):
    return np.maximum(a * x, x)

x = np.arange(-5.0, 5.0, 0.1)
y = leaky_relu(x)

plt.plot(x, y)
plt.plot([0,0], [5.0, 0.0], ':') # 가운데 점선 추가
plt.title('Leaky ReLU Function')
plt.show()
```

```python
# 소프트맥스 함수(Softmax funtion)

# 분류 문제를 로지스틱 회귀와 소프트맥스 회귀를 출력층에 적용하여 사용
# 세 가지 이상의 (상호 베타적인) 선택지 중 하나를 고르는 다중 클래스 분류에 주로 사용

x = np.arange(-5.0, 5.0, 0.1)
y = np.exp(x) / np.sum(np.exp(x))

plt.plot(x, y)
plt.title('Softmax Funtion')
plt.show()
```
